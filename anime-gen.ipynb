```python
from diffusers import DiffusionPipeline
import torch

pipe = DiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,
    variant="fp16",
    use_safetensors=True
).to("cuda")

pipe.enable_model_cpu_offload()
```

```python
# Loading Trained LoRA Weights
pipe.load_lora_weights("caemuller/anime_gen")
```

```python
type(pipe)
```

```python
from PIL import Image
```

```python
init_image = Image.open("foto.jpg")

STRENGTH = 0.3
STEPS=25

negative_prompt="multiple people"

prompt = "cute boy, dark brown hair, dark brown eyes, black shirt"
image = pipe(
    init_image=init_image,
    prompt=prompt,
    num_inference_steps=STEPS,
    height=1024,
    width=1024,
    strenght=STRENGTH,
    guidance_scale=7.5,
    negative_prompt=negative_prompt
).images[0]
```

```python
image.save(f"results/with_context/{prompt}_{STEPS}_steps_{STRENGTH}.png")
```
